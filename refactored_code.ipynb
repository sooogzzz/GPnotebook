```python
# Squared exponential covariance function
def Kse(hyps,x1,x2, inc_amp=True):
    '''
    Covariance matrix calculation with some vectorisation techniques
    
    Parameters:
    ----------
    hyps : Numpy array/list
        Array of hyperparameters, hyps[0] - amplitude, hyps[1:] - length scales
    x1 : Numpy array
        Array of points to evaluate covariance against x2
    x2 : Numpy array
        Array of points to evaluate covariance against x1
    '''
    n1 = x1.shape[0]
    n2 = x2.shape[0]
    m = x1.shape[1]
 
    if inc_amp:
        a = float( hyps[0] )
        ds = hyps[1:]
    else:
        ds = hyps
    
    Ksum = np.zeros( (n1,n2) )

    for i in range(0,m):

        Kdiff = np.tile( x1[:,i] , (n2,1) ).T - np.tile( x2[:,i] , (n1,1) )
        
        Ksum += float( -0.5 * ds[i] ) * np.square(Kdiff,dtype=float) 
    
    K = a * np.exp(Ksum) if inc_amp else np.exp(Ksum, dtype = float)

    return K

# Matern 3/2 covariance function
def Km3_2(hyps,x1,x2, inc_amp=True):
    '''
    Covariance matrix calculation with some vectorisation techniques
    
    Parameters:
    ----------
    hyps : Numpy array/list
        Array of hyperparameters, hyps[0] - amplitude, hyps[1:] - length scales
    x1 : Numpy array
        Array of points to evaluate covariance against x2
    x2 : Numpy array
        Array of points to evaluate covariance against x1
    '''
    n1 = x1.shape[0]
    m = x1.shape[1]
    n2 = x2.shape[0]
    
    if inc_amp:
        a = float( hyps[0] )
        ds = hyps[1:]
    else:
        ds = hyps

    Ksum1 = np.zeros( (n1,n2) )
    Ksum2 = np.zeros( (n1,n2) )

    for i in range(0,m): 

        Kdiffabs = np.abs(np.tile( x1[:,i] , (n2,1) ).T - np.tile( x2[:,i] , (n1,1) ), dtype = float)
        
        Ksum1 += float( -np.sqrt(3)*ds[i] ) * Kdiffabs
         
        Ksum2 += float(np.sqrt(3) * ds[i] ) * Kdiffabs
    
    K = a* (1 + Ksum2) * np.exp( Ksum1, dtype = float ) if inc_amp else (1 + Ksum2) * np.exp( Ksum1, dtype = float )
    
    return K

# Matern 5/2 covariance function
def Km5_2(hyps,x1,x2, inc_amp=True):
    '''
    Covariance matrix calculation with some vectorisation techniques
    
    Parameters:
    ----------
    hyps : Numpy array/list
        Array of hyperparameters, hyps[0] - amplitude, hyps[1:] - length scales
    x1 : Numpy array
        Array of points to evaluate covariance against x2
    x2 : Numpy array
        Array of points to evaluate covariance against x1
    inc_amp: Bool
        True if amplitued hyperparameter is included in hyps
    '''
    n1 = x1.shape[0]
    m = x1.shape[1]
    n2 = x2.shape[0]
    
    if inc_amp:
        a = float( hyps[0] )
        ds = hyps[1:]
    else:
        ds = hyps

    Ksum1 = np.zeros( (n1,n2) )
    Ksum2 = np.zeros( (n1,n2) )
    Ksum3 = np.zeros( (n1,n2) )

    for i in range(0,m): 

        Kdiffabs = np.abs(np.tile( x1[:,i] , (n2,1) ).T - np.tile( x2[:,i] , (n1,1) ), dtype=float)

        Ksum1 += float( -np.sqrt(5)*ds[i] ) * Kdiffabs
        
        Ksum2 += float( (5*ds[i]**2)/3 ) * np.square(Kdiffabs, dtype=float) 
        
        Ksum3 += float(np.sqrt(5) * ds[i] ) * Kdiffabs
    
    K = a*(1 + Ksum3 + Ksum2) * np.exp( Ksum1 , dtype= float) if inc_amp else (1 + Ksum3 + Ksum2) * np.exp( Ksum1 , dtype= float)
    
    return K

def Krq(hyps,x1,x2, inc_amp = True):
    '''
    Covariance matrix calculation with some vectorisation techniques
    
    Parameters:
    ----------
    hyps : Numpy array/list
        Array of hyperparameters, hyps[0] - amplitude, hyps[1] - inverse of scale mixture parameter, hyps[2:] - reciprical length scales
    x1 : Numpy array
        Array of points to evaluate covariance against x2
    x2 : Numpy array
        Array of points to evaluate covariance against x1
    inc_amp: Bool
        True if amplitued hyperparameter is included in hyps
    '''
    n1 = x1.shape[0]
    m = x1.shape[1]
    n2 = x2.shape[0]
    
    if inc_amp:
        a = float( hyps[0] )
        omega = float(hyps[1])
        ds = hyps[2:]
    else:
        omega = float(hyps[0])
        ds = hyps[1:]
        
    Ksum = np.zeros( (n1,n2) )

    for i in range(0,m): 

        Kdiff = np.tile( x1[:,i] , (n2,1) ).T - np.tile( x2[:,i] , (n1,1) )
        
        Ksum += 0.5 * omega * float(ds[i]) * np.square(Kdiff,dtype=float) 
    
    K = a * (1 + Ksum)**(-omega) if inc_amp else (1 + Ksum)**(-omega)
    
    return K

# derivative of SE w.r.t d hyper-parameter
def dKse(ds, x, i):
    '''
    First order derivative of covariance function w.r.t. length scale parameter with some vectorisation techniques
    
    Parameters:
    ----------
    Kfunc : function
        Associated covariance function
    ds : float
        Length scale parameters
    x : Numpy array
        Domain
    i : integer
        Differentiate w.r.t. the i-th hyperparameter
    '''
    n = x.shape[0]
    m = x.shape[1]
    
    Kdiff = np.tile( x[:,i] , (n,1) ).T - np.tile( x[:,i] , (n,1) )
    Ksum = np.square(Kdiff,dtype=float)

    DdQ = -float( ds[i] )/2 * Ksum * Kse(ds, x, x, inc_amp = False) # pointwise multiplication
    
    return DdQ


# derivative of Matern 3/2 covariance function w.r.t d hyper-parameter
def dKm3_2(ds, x, i):
    '''
    First order derivative of covariance function w.r.t. length scale parameter with some vectorisation techniques
    
    Parameters:
    ----------
    ds : float
        Length scale parameters
    x : Numpy array
        Domain
    i : integer
        Differentiate w.r.t. the i-th hyperparameter
    '''
    n = x.shape[0]
    m = x.shape[1]

    Kdiffi = np.abs(np.tile( x[:,i] , (n,1) ).T - np.tile( x[:,i] , (n,1) ), dtype=float)
    Ksum1i = 3*float(ds[i])**2*np.square(Kdiffi, dtype=float)
    
    Ksum2 = np.zeros( (n,n) )

    for j in range(0,m):
        Kdiff = np.abs(np.tile( x[:,j] , (n,1) ).T - np.tile( x[:,j] , (n,1) ), dtype=float) if j != i else Kdiffi
        Ksum2 += -np.sqrt(3)*float(ds[i])*Kdiff
    
    DdQ = -Ksum1i*np.exp( Ksum2, dtype = float)
    
    return DdQ

# derivative of Matern 5/2 covariance function w.r.t d hyper-parameter
def dKm5_2(ds, x, i):
    '''
    First order derivative of covariance function w.r.t. length scale parameter with some vectorisation techniques
    
    Parameters:
    ----------
    ds : float
        Length scale parameters
    x : Numpy array
        Domain
    i : integer
        Differentiate w.r.t. the i-th hyperparameter
    '''
    n = x.shape[0]
    m = x.shape[1]
    
    Kdiffi = np.abs(np.tile( x[:,i] , (n,1) ).T - np.tile( x[:,i] , (n,1) ), dtype=float)
    Ksum1i = 5*float(ds[i])**2/3 * np.square(Kdiffi,dtype = float)
    Ksum2i = 1 + np.sqrt(5)*float(ds[i])*Kdiffi

    Ksum3 = np.zeros( (n,n) )
    
    for j in range(0,m):
        Kdiff = np.abs(np.tile( x[:,j] , (n,1) ).T - np.tile( x[:,j] , (n,1) ), dtype=float) if j != i else Kdiffi

        Ksum3 += -np.sqrt(5)*float(ds[i])*Kdiff
        
    DdQ = -Ksum1i*Ksum2i*np.exp(Ksum3,dtype=float) 
    
    return DdQ

# derivative of Rational quadratic covariance function
## needs to be checked
def dKrq(ds, omega, x, i, wrt_omega = False):
    '''
    First order derivative of covariance function w.r.t. reciprical length scale parameter d with some vectorisation techniques
    
    Parameters:
    ----------
    K : Numpy array
        Covariance matrix evaluated on domain x
    d : numpy array or float
        Length scale parameter array or float
    omega : float
        Inverse scale mixture parameter
    x : Numpy array
        Domain
    '''
    n = x.shape[0]
    m = x.shape[1]
    hyps = [omega] + ds
    
    
    if wrt_omega:
        Ksum1 = np.zeros( (n,n) )
        Ksum2 = np.zeros( (n,n) )
        
        for j in range(m):
            Kdiff = np.tile( x[:,j], (n,1) ).T - np.tile( x[:,j] , (n,1) )
            Ksquare = np.square(Kdiff,dtype=float)
            Ksum1 += np.log(1+Ksquare*omega* float( ds[j]), dtype=float) 
            Ksum2 += omega*Ksquare* float( ds[j])/(1+Ksquare*omega* float( ds[j]))

        DdQ = (Ksum1-Ksum2)*Krq(hyps,x,x, inc_amp = False)/omega

        return DdQ
    else:
        Kdiff = np.tile( x, (n,1) ).T - np.tile( x , (n,1) )
        Ksquare = np.square(Kdiff,dtype=float)
        Ksum1 = Ksquare*ds[i]
        Ksum2 = 1 + Ksquare*ds[i]*omega

        DdQ = -Ksum1*Krq(hyps,x,x)/Ksum2
        
        return DdQ
```

```python
# Negative log-likelihood function (constants and scaling removed)
def nLL(log_hyps,Kfunc,dKfunc,x,y, inc_noise):
    '''
    Calculates the negative log likelihood given kernel func, hyperparameters, inputs x and outputs y
    
    Parameters:
    ----------
    log_hyps : Numpy array/list of floats
        Array of hyperparameters, length depends on func. If inc_noise, log_hyps[-1] - noise variance
    K_func : function
        Covariance function to calculate the nLL
    dK_func : function
        Derivative of covariance function with respect to hyperparameters
    x : Numpy array
        Array of points in the input domain
    y : Numpy array
        Outputs corresponding to the input points
    inc_noise : bool
        If true, include noise hyperparameter
    '''

    if inc_noise:
        hypsK = np.exp ( log_hyps[:-1] )
        v =  np.exp( log_hyps[-1] ) 
        
        K = Kfunc(hypsK,x,x) 
        
        # Add noise variance
        Q = diagAdd(v,K)
    else:
        hypsK = np.exp ( log_hyps )
        # print('nll: ' + str(hypsK))
        Q = Kfunc(hypsK, x, x) 
    

    # invert
    invQ = scipy.linalg.inv(Q)
    
    # Numerically stable calculation  of logdet(Q)
    log_det = np.sum( np.log ( scipy.linalg.svd(Q)[1]) ) 

    negLogLike = float( np.transpose(y) @ invQ @ y + log_det )
    
    
    return negLogLike

# nLL Jacobian function
def DnLL(log_hyps,Kfunc, dKfunc,x,y, inc_noise):
    '''
    Calculates the gradient of the negative log likelihood given kernel function, hyperparameters, inputs x and outputs y
    
    Parameters:
    ----------
    log_hyps : Numpy array/list of floats
        Array of hyperparameters, length depends on func. If inc_noise, log_hyps[-1] - noise variance
    K_func : function
        Covariance function to calculate the nLL
    dK_func : function
        Derivative of covariance function with respect to hyperparameters
    x : Numpy array
        Array of points in the input domain
    y : Numpy array
        Outputs corresponding to the input points
    inc_noise : bool
        If true, include noise hyperparameter
    '''
    n_hyps = len(log_hyps)
    n = x.shape[0]
    m = x.shape[1]
    if inc_noise:
        hypsK = np.exp ( log_hyps[:-1] )
        v = float( np.exp( log_hyps[-1] ) )
        
        K = Kfunc(hypsK,x,x) 
    
        # Add noise variance and invert
        Omega = diagAdd( v , np.matlib.zeros( (n,n) ) )
        Q = K + Omega 

    else:
        hypsK = np.exp ( log_hyps )

        K = Kfunc(hypsK,x,x)
        Q = K
        
    invQ = scipy.linalg.inv(Q)
    yTinvQ = np.transpose(y) @ invQ
    
    Jac = np.zeros( ( n_hyps ,) )
    DaQ = K
    Jac[0] = traceAB(invQ,DaQ) - yTinvQ @ DaQ @ yTinvQ.transpose()
    
    a = float(hypsK[0])
    ds = hypsK[1:(m+1)] 

            
    for i in range(0,m):
        DdQ = a*dKfunc(ds, x, i)

        Jac[i+1] = traceAB(invQ,DdQ) - yTinvQ @ DdQ @ yTinvQ.transpose()
        
    if inc_noise:
        DvQ = Omega
        Jac[m + 1] = traceAB(invQ,DvQ) - yTinvQ @ DvQ @ yTinvQ.transpose()

        
    return Jac

# Perform hyperparameter learning
def hypLearn_exp(Kfunc, dKfunc,x,y, log_hyps0, inc_noise, optimizer = 'BFGS'):
    '''
    Perform MLE to estimate the hyperparameters from data. 
    Uses the scipy.optimize minimise method to minimise the negloglike.
    
    Parameters:
    ----------
    K_func : function
        Covariance function to calculate the nLL
    dK_func : function
        Derivative of covariance function with respect to hyperparameters
    x : Numpy array
        Array of points in the input domain
    y : Numpy array
        Outputs corresponding to the input points
    n_hyps : int
        Number of hyperparameters
    inc_noise : bool
        If true, include noise hyperparameter
    '''

    res = minimize( nLL , log_hyps0, (Kfunc,dKfunc,x,y, inc_noise), method = optimizer, jac = DnLL, options = {'disp': True}  )
    hyps = np.exp( res.x )
    print('Learned hyperparameters [a,d,v] = ', hyps) if inc_noise else print('Learned hyperparameters [a,d] = ', hyps)
    
    return hyps
```

```python
def calc_rmse(actual,predicted):  
    return np.sqrt(np.square(np.subtract(actual,predicted)).mean() )

def CountUpcross(y):
    return np.sum([cross for cross in (y[1:] > 0) & (y[:-1] <= 0)])  

# Trace(A*B) function to save computational time
def traceAB(A,B): 
    '''
    Calculates trace of A*B
    '''
    # A and B assumed to be square
    n = A.shape[0]

    trace = float(0)

    for i in range(0,n):
        trace += float( A[i,:] @ B[:,i] ) 
    
    return trace

# diag-add function
def diagAdd(v,A):
    '''
    Adds scalar value to the diagonal of A
    '''
    # n = A.shape[0]
    B = np.copy(A)
    np.fill_diagonal(B, B.diagonal() + v) 
    

    # for i in range(0,n):
        
    #     A[i,i] += v

    return B

# GP predictions
def GPpred(K_func,hyps,x,y,x_pred):
    '''
    Make predictions on input domain x_pred.
    '''

    Kx = K_func(hyps[:-1],x,x)
    Q = diagAdd(hyps[-1],Kx)
    invQ = la.inv(Q)

    Kpx = K_func(hyps[:-1],x_pred,x)
    Kpp = K_func(hyps[:-1],x_pred,x_pred)
    
    pred = Kpx @ invQ @ y # GP prediction on x_pred

    cov = Kpp - Kpx @ invQ @ Kpx.transpose() # GP covariance matrix on x_pred
    std = np.sqrt( np.diag( cov ) ).reshape(-1,1) # GP standard deviations on x_pred

    return pred, std
```

```python
def testRun3(x = None, y_sample = None, y_noise = None):
    '''
    One dimensional case, learning two hyperparameters from a sample of the GP prior with added noise.
    Plotting the negative log likelihood against values of d with exact values of other hyps.
    '''
    ## take sample if not provided
    if y_sample == None:
        n = 200
        xmin = 1
        xmax = 20
        x = np.linspace(xmin,xmax,n).reshape(-1,1)

        # define hyps
        a = 1
        d = 2

        Kx = Kse([a,d],x,x)
        sqrtKx = np.real( la.sqrtm(Kx) )

        sig = 0.2
        y_sample = (np.random.normal(0,1, (1,n))@sqrtKx).reshape(-1,1) 
        y_noise = y_sample + np.random.normal(0,sig, (n,1) )  
    
    # Hyperparameter training
    log_hyps0 = np.random.normal(0,1, (3,) )
    hyps = hypLearn_exp(Kse, dKse, x, y_noise, log_hyps0, inc_noise=True)
    loglike_opt = nLL(np.log(hyps),Kse, dKse, x,y_noise, inc_noise=True)
    
    # calculate negative loglikelihood function
    phi_a = np.log(hyps[0])
    phi_v = np.log(hyps[2])
    loglike = []
    d_ar = np.linspace(0.5,4,20)
    loglike_act = nLL([phi_a, np.log(d), phi_v],Kse, dKse, x,y_noise, inc_noise=True)
    phi_d_ar = np.log(d_ar)
    for phi_d in phi_d_ar:
        log_hyps = [phi_a, phi_d, phi_v]
        loglike += [nLL(log_hyps,Kse, dKse, x,y_noise, inc_noise=True)]

    
    plt.figure(figsize=(10, 8), dpi=80, facecolor=(1,1,1))
    plt.plot(d_ar,loglike, color = 'red',label = r'\textbf{Loglike}', alpha = 1, linewidth = 2)
    plt.plot(hyps[1],loglike_opt,'bo', label = 'Minimum'),
    plt.plot(d,loglike_act,'go', label = 'Correct'),
    plt.xlabel(r'\boldmath$d$')
    plt.ylabel(r'\textbf{Negative Loglike Value}')
    plt.legend()
    plt.show()

    return

# run test code
testRun3()#x = x, y_sample = y,y_noise=y_noise)
```
